```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Постройте график, Назовите график, Сделайте именование оси x и оси y. Сделайте выводы

```python
df['price'].min()
```

1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(15, 6))
sns.histplot(df['price'])
plt.ticklabel_format(style='plain')
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Кол-во');
```

1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6, 8))
sns.histplot(df['sqft_living'], bins=50)
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество');
```

1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(10, 15))
sns.histplot(df['yr_built'], bins=40)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

Условие 2: 2 задача

2.1. Изучите распределение домов от наличия вида на набережную


Постройте график. Сделайте выводы

```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['no', 'yes']);
```

2.2. Изучите распределение этажей домов

```python
data2 = df['floors'].value_counts()
data2
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x = data2.index, y = data2)
plt.title('Распределение этажей домов')
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во')
plt.xticks(rotation=30);
```

2.3. Изучите распределение состояния домов

```python
data3 = df['condition'].value_counts()
data3
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x = data3.index, y = data3)
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Кол-во');
```

Условие 3: 3 задача


Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
sns.jointplot(x=df['цена'], y=df['sqft_living'], kind='reg')
sns.jointplot(x=df['цена'], y=df['sqft_above'], kind='reg')
sns.jointplot(x=df['цена'], y=df['grade'], kind='reg')
sns.jointplot(x=df['цена'], y=df['bedrooms'], kind='reg')
sns.jointplot(x=df['цена'], y=df['bathrooms'], kind='reg')
sns.jointplot(x=df['цена'], y=df['yr_built'], kind='reg');
```

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('цена')
plt.ylabel('вид')
plt.title('Распределение цены по видам');
```
